```python
from gensim.models import Word2Vec

# Training data: tokenized sentences
sentences = [["machine", "learning"], ["deep", "learning"], ["artificial", "intelligence"]]

# Create the model
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

# Get vector for a word
print(model.wv["machine"])  # Outputs a 100-dimensional vector
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(["This is an example sentence."])

print(embeddings)  # 384-dimensional vector for the sentence
```
